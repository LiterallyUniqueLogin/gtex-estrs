```python
#ch='chr21'; gene='ENSG00000160213.5' ; var_str=45196326 ; var_snp=45214684

import statsmodels.formula.api as smf
import statsmodels.api as sm
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

Data = pd.read_csv("/storage/szfeupe/Runs/GTEx_estr/Analysis_by_Tissue/WholeBlood/Master.table", sep='\t')

Data = pd.read_csv("/storage/szfeupe/Runs/GTEx_estr/Analysis_by_Tissue/SUMMARY/Merged_Causality_best_tissue.table", sep='\t')
data = Data.sort_values('best.score', ascending=[False])
data = data.head(20)
Gene_table = pd.read_csv('/storage/szfeupe/Runs/GTEx_estr/FEATURES/Genes_only_table', sep='\t')


chs = list(data['chrom'])
genes = list(data['gene'])
starts = list(data['best.str.start'])
tissues = list(data['best.tissue'])
scores = list(data['best.score'])

for i in range(11):
    try: 
        print ('\n....',i,'....',str(tissues[i]))
    except:
        print ('PASS************')
        continue
    print('------------->',chs[i],' gene ',genes[i],'\n')
    snps=pd.read_csv("/storage/szfeupe/Runs/GTEx_estr/Analysis_by_Tissue/"+tissues[i]+"/SNP_Analysis/PQValues.txt", sep="\t")
    snp =pd.read_csv("/storage/szfeupe/Runs/GTEx_estr/SNP_Analysis/SNP_Geno"+chs[i], sep='\t') #snps
    Genos=pd.read_csv("/storage/szfeupe/Runs/GTEx_estr/Normalized_Genotypes/STR_Raw_Gen."+chs[i], sep='\t')
    expr = pd.read_csv("/storage/szfeupe/Runs/GTEx_estr/Analysis_by_Tissue/"+tissues[i]+"/Corr_Expr.csv",sep=',' )
    
    S = snps.loc[snps['gene']==genes[i]]['str.start']; S=int(S)
    Gene = Gene_table.loc[Gene_table['gene']==genes[i]]['gene.name'].values[0]
    
    samples = list(expr.index)
    
    Genosnp = snp.loc[snp['start']==S,samples]
    #print(Genosnp.shape)
    Genostr = Genos.loc[Genos['start']==starts[i], samples]
    Expr = expr[genes[i]]
    
    Genosnp = Genosnp.replace('None', np.nan) ; #print(Genosnp)
    Genostr = Genostr.replace('None', np.nan) ; #print(Genostr)
    #samples = list(expr.index)
    
    plt.figure(figsize=(8, 6))
    plt.subplot(221)
    plt.title(tissues[i]+' tissue')
    plt.scatter(Genosnp, Expr, label='Lead SNP', color='b')
    plt.ylabel(Gene+' expression')
    plt.xlabel('Best SNP genotypes')
    plt.subplot(222)
    plt.scatter(Genostr, Expr,label='eSTR', color='g')
    #plt.ylabel(Gene+' Gene expression')
    plt.xlabel('eSTR genotypes \nSTR_'+str(int(starts[i])))
    plt.show()
 

    ##Hold snp genotypes constant
    geno=list(set(Genosnp.values.tolist()[0]))
    geno=list(set([x for x in np.nan_to_num(geno) if x!='nan']))
    n=len(geno)
    ind=Genosnp.index[0]
    fig = plt.figure(figsize=(8,10)) 
    for i in range(len(geno)):
        print(geno[i])
        S = Genosnp.T.loc[Genosnp.T[ind]==geno[i]]
        sample=list(S.index)
        S['Exp']=Expr.loc[sample]
        S['Str']=Genostr.T.loc[sample][Genostr.index[0]]
        if len(S['Exp'])==0:
            continue
        ax = fig.add_subplot(n,1,1+i)
        ax.scatter(list(S['Str']), list(S['Exp']))
        ax.title.set_text(Gene+' Expression @ SNP_genotype='+str(geno[i])+' ( n='+str(len(S['Exp']))+' )')
    plt.show()

    
```
